```python
clf_name = 'ECOD'
clf = ECOD(contamination=0.4)
```

```python
df_Sztd2_2=pd.DataFrame({"Sztd2_2":df_mini.Szteroid2_2_rhv_hom.diff()})
```

```python
od_pred = clf.labels_  # binary labels (0: inliers, 1: outliers)
od_scores = clf.decision_scores_  # raw outlier scores
```
